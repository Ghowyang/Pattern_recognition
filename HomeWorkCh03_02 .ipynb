```python
import numpy as np
import scipy as sp
import scipy.stats as st
import sympy as sm
import matplotlib.pyplot as pl
from scipy.special import gamma, factorial
```

```python
x = np.linspace(-5, 5, 100)
y = gamma(x)
```

```python
pl.plot(x, y, 'b', alpha=0.8, label='gamma(x)')
pl.xlim(-5, 5)
pl.ylim(-25, 25)
pl.grid()
pl.xlabel('x')
pl.legend(loc='lower right')
pl.show()
```
